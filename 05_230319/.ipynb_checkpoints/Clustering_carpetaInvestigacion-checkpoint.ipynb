```python
#tomar 10000 filas de sólo las dos columnas siguientes
fields=["Colonia","Latitud"]
df = pd.read_csv("../carpetas-de-investigacion-pgj-cdmx.csv", sep=";",usecols = fields, nrows = 10000)
df.head()
```

```python
#agrupar los datos por colonia
group = df.groupby(["Colonia"])
#mostrar cuántos datos hay por colonia
group.count()
```

```python
#dff = pd.get_dummies(df["Colonia"])
#df = df.astype(np.float64)

#tomar dos columnas y quitar datos vacíos
newdf = df[["Colonia", "Latitud"]].dropna()

#newdf = newdf.astype(np.float64)
```

```python
newdf.head()
```

```python
#función numpy para agrupar datos por categoría: en este caso, Colonia
colonias = np.unique(newdf["Colonia"])
#print(len(colonias))

#para cada colonia, almacenar:
#nombre de coloni-índice de colonia
numColonias = []
for i in range(len(colonias)):
    numColonias.append([colonias[i], i])

numColonias
```

```python
#sobreescribir dataframe newdf,
#reemplazar nombre de la colonia por el índice de la colonia
for i in range(len(numColonias)):
    newdf.replace(to_replace = numColonias[i][0], value = numColonias[i][1], inplace = True)

newdf.head()
```

```python
#crear 5 clusters de newdf
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, n_jobs = -1).fit(newdf)

#mostrar las etiquetas de los clusters
kmeans.labels_
```

```python
import matplotlib.pyplot as plt
#eje X: colonia|eje Y: latitud
plt.scatter(newdf["Colonia"], newdf["Latitud"], c = kmeans.labels_)
plt.show()
```
