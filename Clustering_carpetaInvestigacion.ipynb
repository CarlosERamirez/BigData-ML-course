```python
fields=["Colonia","Latitud"]
df = pd.read_csv("carpetas-de-investigacion-pgj-cdmx.csv", sep=";",usecols = fields, nrows = 10000)
df.head()
```

```python
group = df.groupby(["Colonia"])
group.count()
```

```python
#dff = pd.get_dummies(df["Colonia"])
#df = df.astype(np.float64)
newdf = df[["Colonia", "Latitud"]].dropna()
#newdf = newdf.astype(np.float64)
```

```python
newdf.head()
```

```python
colonias = np.unique(newdf["Colonia"])
#print(len(colonias))
numColonias = []
for i in range(len(colonias)):
    numColonias.append([colonias[i], i])

#numColonias
```

```python
for i in range(len(numColonias)):
    newdf.replace(to_replace = numColonias[i][0], value = numColonias[i][1], inplace = True)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, n_jobs = -1).fit(newdf)
kmeans.labels_
```

```python
import matplotlib.pyplot as plt
plt.scatter(newdf["Colonia"], newdf["Latitud"], c = kmeans.labels_)
plt.show()
```
